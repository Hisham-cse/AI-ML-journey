```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load dataset
df = pd.read_csv("salary_data.csv")   # columns: YearsExperience, Salary
```

```python

# Basic Exploration
print(df.head())

```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
# Scatter Plot
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()

```

```python
print(df.isnull().sum())
```

```python
# Correlation Heatmap
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
