```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5])
y = np.array([30000, 35000, 40000, 45000, 50000])

m = 0
c = 0
lr = 0.01
n = len(X)

for i in range(1000):
    y_pred = m * X + c
    dm = (-2/n) * sum(X * (y - y_pred))
    dc = (-2/n) * sum(y - y_pred)
    m = m - lr * dm
    c = c - lr * dc

print("Slope:", m)
print("Intercept:", c)

plt.scatter(X, y)
plt.plot(X, m * X + c, color="red")
plt.show()
```
