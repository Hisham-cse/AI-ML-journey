```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([30000, 35000, 40000, 45000, 50000])
```

```python

# Initial values
m = 0
b = 0
learning_rate = 0.01
n = len(X)
epochs = 1000
```

```python

# Training using Gradient Descent
for i in range(epochs):
    y_pred = m * X + b

    # Derivatives
    dm = (-2/n) * sum(X * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # Print loss occasionally
    if i % 100 == 0:
        loss = sum((y - y_pred) ** 2) / n
        print(f"Epoch {i} | Loss: {loss:.2f}")

print("\nFinal parameters:")
print("Slope (m):", m)
print("Intercept (b):", b)
```

```python

# Final prediction line
y_final = m * X + b
y_final
```

```python

# Plot
plt.scatter(X, y, color="blue")
plt.plot(X, y_final, color="red")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Using Gradient Descent (From Scratch)")
plt.show()
```
